# Harmonic mean (H.M) of systems:
- The harmonic mean is a type of numerical average. It is calculated by dividing the nu,mber of observations by the resprocal of each number in the series. Thus, the harmonic mean is the resprocal of the artimathic mean of the resprocals. 

## $H.M=\frac{n}{\sum(\frac{1}{x_{n}})}$

```python
print ('H.M-Porb:  %0.4f' % stats.hmean(e1)) 
print ('H.M-q   :  %0.4f' % stats.hmean(e2))
print ('H.M-Ma  :  %0.4f' % stats.hmean(e3))
print ('H.M-Md  :  %0.4f' % stats.hmean(e4))
print ('H.M-M   :  %0.4f' % stats.hmean(e5))
print ('H.M-a   :  %0.4f' % stats.hmean(e6))
print ('H.M-r_r:   %0.4f' % stats.hmean(e7))
print ('H.M-r_L :  %0.4f' % stats.hmean(e8))
print ('H.M-beta:  %0.4f' % stats.hmean(e9))
print ('H.M-q_cr:  %0.4f'  %stats.hmean(e10))
```

# Geometric mean (G.M) of systems

## $G.M=(x_{1}.x_{2}...x_{n})^{1/n}$

## $\bar{x}(A.M)\geq(G.M)\geq(H.M)$

- The Geometric mean is used in finance to calculate average groth rates and is referred to as the 
compounded as annual rate. Consider a stock that grows by 10\% in a year, declines by 20\% in year two, 
and then, growth by 30\% in year three...etc.

- Geometry is used in Astronomy to measure the speed, velocity, momentum, direction, distance, realtivity and probability

```python
print('G.M-Porb:  %0.4f' % stats.gmean(e1)) 
print('G.M-q   :  %0.4f' % stats.gmean(e2))
print('G.M-Ma  :  %0.4f' % stats.gmean(e3))
print('G.M-Md  :  %0.4f' % stats.gmean(e4))
print('G.M-M   :  %0.4f' % stats.gmean(e5))
print('G.M-a   :  %0.4f' % stats.gmean(e6))
print('G.M-r_r:   %0.4f' % stats.gmean(e7))
print('G.M-r_L :  %0.4f' % stats.gmean(e8))
print('G.M-beta:  %0.4f' % stats.gmean(e9))
print('G.M-q_cr:  %0.4f' %stats.gmean(e10))
```
