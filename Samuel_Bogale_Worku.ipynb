```python
def factorial(n):
    "Computes and returns the factorial of n, a positive integer."
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
print(factorial(1))
print(factorial(2))
print(factorial(3))
print(factorial(4))
print(factorial(5))
print(factorial(6)) 
print(factorial(7))
```

```python
def fibonacci(n):
    "Computes and returns the Fibonacci of n, a positive integer."
    if n == 1:
        return 1
    elif n == 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
```

```python
print(fibonacci(3))
print(fibonacci(4))
print(fibonacci(5))
print(fibonacci(6))
print(fibonacci(7))
print(fibonacci(8))
print(fibonacci(9))
print(fibonacci(10))
print(fibonacci(11))
```

##  $$\frac{dm}{dr}=4\pi\rho r^{2}$$

## Mass  Conservation

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from numpy import  pi
```

```python
def model(m,r):
   
    rho=1.41
    dmdr = 4 *pi * r**2 * rho
    return dmdr
m0 =1.99 # Mass of the Sun in 10^30

r=np.linspace(1.5, 6.8, 10)

# solve ODE
m = odeint(model,m0,r)
#print(m)
plt.figure(figsize=(14,6))
plt.plot(r,m,'b',label='radius vs mass', linewidth=4)
plt.xlabel('  $(R_{\odot})}$  in $meter$ \n interms  of solar Radius ',fontsize=27)
plt.ylabel('   $(M_{\odot})}$ in Kg \n interms  of solar Mass ',fontsize=27)
plt.legend(loc=2, fontsize =24)
plt.title('Mass Conservation',fontsize=24)
plt.rc('xtick',labelsize=22)
plt.rc('ytick',labelsize=22)

plt.show()
```

###  $$\frac{dP}{dr} = \frac {-GMr\rho} {r}^{2}$$

## Hydrodynamic Equiliburim

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def model(p,r):
    G=6.67e-11
    rho=1.41
    M=1.99e+30
    dpdr =-G * M * rho**2 / r
    return dpdr
p0 =5

r=np.linspace(1.5, 7.6, 10)
p = odeint(model,p0,r)

plt.figure(figsize=(14,6))
plt.plot(r,p, 'r',label='Pressure vs radius', linewidth=5 )
plt.xlabel('$(R_{\odot})}$ in $mete$ \n interms  of solar Radius '  ,fontsize=27)
plt.ylabel('$(P_{\odot})}$ in passcal  \n interms  of solar Pressure '  ,fontsize=27)
plt.legend(loc=1, fontsize =24)
plt.title('HydroStatics Equilbrium',fontsize=24)
plt.rc('xtick',labelsize=22)
plt.rc('ytick',labelsize=22)
plt.show()
```

##  Energy Equation

###   $$\frac{dL}{dr}=4\pi r^{2} \rho \epsilon $$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from numpy import pi
```

```python
def model(L,r):
    
    epsilon= 2.5
    rho=1.41
    dLrdr = 4 * pi* r**2  *  epsilon * rho
    return dLrdr


L0 =1.00  ## Luminosity of our Sun

r=np.linspace(1.5, 6.8, 10)
L = odeint(model,L0,r)

plt.figure(figsize=(14,6))
plt.plot(r,L,'g',label='Luminosity  vs radius', linewidth=5)
plt.xlabel(' $(R_{\odot})}$  $meter$ \n interms  of solar Radius ' ,fontsize=27)
plt.ylabel(' $(L_{\odot})}$ in watt \n interms  of Luminosity of the sun ',fontsize=27)
plt.legend(loc=2, fontsize =24)
plt.title('Energy Transfer Equation',fontsize=24)
plt.show()
```

### $$ \frac{dT}{dr} =\frac{-3}{4ac} \frac{kp} {t^{3}} \frac{L}{4\pi r^{2}} $$

## Radiative tranport  or Temperature gradient

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def model(T,r):
    a=0.5
    c=1.0
    k=-6.50
    p=7.50
    l=1.00
    dTdt=(-3/4*a*c)*(k*p)/(T**3) *(l/(4*pi)*(r**2))
    return dTdt
T0=5.8    # Temperature of our sun
ro=6.98    # Radius of our Sun
r=np.linspace(1.5, 6.8, 10)
T = odeint(model,T0,r)

plt.figure(figsize=(14,6))
plt.plot(r,T,'m',label='Temperature vs radius ', linewidth=5)
plt.xlabel('  $(R_{\odot})}$ in $meter$ \n interms  of solar Radius',fontsize=27)
plt.ylabel(' $(T_{\odot})}$ $ ^{o}k$ \n interms  of solar Temperature',fontsize=27)
plt.legend(loc=2, fontsize =24)
plt.title('Temperature gradiant',fontsize=24)
plt.rc('xtick',labelsize=22)
plt.rc('ytick',labelsize=22)
plt.show()
```
